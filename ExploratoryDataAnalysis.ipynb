```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
penguins = sns.load_dataset('penguins')
penguins
```

```python
# Fill in NaN values with the species-specific mean or randomly
sex_value=('Female', 'Male')
for column in penguins.columns:
  if penguins[column].dtype == 'float':
    species_means = penguins.groupby('species')[column].mean().to_dict()
    penguins[column].fillna(penguins['species'].map(species_means), inplace=True)
  else:
    penguins['sex'].fillna(np.random.choice(sex_value),inplace=True)
penguins
```

# 2/ Chia dataset thành tập train và tập test với random_state khi sử dụng hàm train_test_split là 22520004


```python
# Convert categorical variables ('species,' 'island,' and 'sex') into numerical labels
from sklearn.preprocessing import LabelEncoder
penguins["species"] = LabelEncoder().fit_transform(penguins["species"])
penguins["island"] = LabelEncoder().fit_transform(penguins["island"])
penguins["sex"] = LabelEncoder().fit_transform(penguins["sex"])
```

```python
# Split the data into training set and test set
from sklearn.model_selection import train_test_split
X = penguins[["island", "bill_length_mm", "bill_depth_mm", "flipper_length_mm",	"body_mass_g", "sex"]]
y = penguins["species"] # Targets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=22520004)
```

```python
# Train DecisionTree model
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth=4, random_state=42)
model.fit(x_train, y_train)
```

```python
# Predict on the test data and measure accuracy
from sklearn.metrics import accuracy_score
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy on test set:", accuracy)
```

# 3/ Thử một số setting khác nhau cho hyperparameter 'max_depth' và nhận xét về sự ảnh hưởng tới độ chính xác của mô hình trên tập test

```python
all_depth_values = list(range(0, 21))
test_accuracy = []

# Measure accuracy with max_depth from 'None' to 20
for max_depth in range(21):
  if max_depth == 0:
    depth_value = None
  else:
    depth_value = max_depth
  model = DecisionTreeClassifier(max_depth=depth_value, random_state=42)
  model.fit(x_train, y_train)
  y_test_pred = model.predict(x_test)
  test_accuracy.append(accuracy_score(y_test, y_test_pred))

# Draw graph showing the relationship between max_depth and accuracy on the test set
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(all_depth_values, test_accuracy, label='Accuracy on the test set', marker='o')
plt.xlabel('max_depth')
plt.ylabel('accuracy')
plt.title('Biểu đồ thể hiện mối quan hệ giữa max_depth với độ chính xác của mô hình trên tập test')
plt.xticks(all_depth_values)
plt.yticks(test_accuracy)
plt.legend()
plt.grid(True)
plt.show()
```

**Nhận xét về sự ảnh hưởng của max_depth tới độ chính xác trên tập test của mô hình Decision Tree:**

- Khi `max_depth = 1`, độ chính xác trên tập test khá thấp (0.8173). Mô hình chưa thể hiểu và mô tả mối quan hệ phức tạp trong dữ liệu.

- Khi `max_depth = 2`, độ chính xác tăng lên đáng kể, đạt 0.9712. Mô hình đã có khả năng tổng quát hóa tốt hơn, cho thấy sự cải thiện rõ so với `max_depth = 1`.

- Khi `max_depth >= 3 cho đến None (không giới hạn độ sâu)`, độ chính xác trên tập test cao nhất, đạt 0.9808. Tuy nhiên, việc tăng max_depth sau đó không còn ảnh hưởng đến độ chính xác của mô hình nữa.

Tổng quan, hyperparameter max_depth có ảnh hưởng lớn đến độ chính xác trên tập test của mô hình Decision Tree. Chọn giá trị max_depth tốt sẽ cho độ chính xác cao. Tuy nhiên, sau một ngưỡng nhất định (trong trường hợp này là max_depth = 3), việc tăng max_depth không cải thiện độ chính xác nữa.
